```python
import pandas as pd
vgsales_inf= pd.read_csv('./vgsales.csv')
vgsales_inf.info()
```

```python
vgsales_inf.head()
```

**Which company is the most common video game publisher?**

```python
most_common_publisher  = vgsales_inf['Publisher'].value_counts().idxmax()
most_common_publisher 
```

**What’s the most common platform?**

```python
most_common_platform = vgsales_inf['Platform'].value_counts().idxmax()
most_common_platform
```

**What about the most common genre?**

```python
most_common_genre = vgsales_inf['Genre'].value_counts().idxmax()
most_common_genre
```

**What are the top 20 highest grossing games?**

```python
top_twenty_highest_grossing_games = vgsales_inf[["Name","Global_Sales"]].sort_values('Global_Sales', ascending=False).head(20)
top_twenty_highest_grossing_games
```

**For North American video game sales, what’s the median?**

```python
na_median_sales=vgsales_inf[['NA_Sales']].median()
na_median_sales
```

**assume that games with same median value are sorted in descending order**

```python
# highly_ten_games =  vgsales_inf[vgsales_inf['NA_Sales'] == median_game_sales].head(10)
# highly_ten_games

# median_game_sales=vgsales_inf[['NA_Sales']].median()
# median_game_sales
# highly_ten_games=vgsales_inf[(vgsales_inf['NA_Sales']>=median_game_sales) & (vgsales_inf['NA_Sales']<=median_game_sales)].head(10)
# highly_ten_games

median_game_sales = vgsales_inf['NA_Sales'].median()
median_game_sales

ten_median_na_seller_names=vgsales_inf[vgsales_inf['NA_Sales']==median_game_sales].head(10)
ten_median_na_seller_names
```

**Top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?**

```python
std_data=vgsales_inf['NA_Sales'].std()
mean_data=vgsales_inf['NA_Sales'].mean()
selling_game=vgsales_inf['NA_Sales'] .head(1)
z_score =( (selling_game-mean_data)/std_data)
z_score
```

**The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?**

```python
# wii_sales=vgsales_inf[['Platform']=='Wii']
# wii_sales_mean=wii_sales.Global_Sales.mean()
# wii_sales_mean

group_by_wii=vgsales_inf[vgsales_inf['Platform']=='Wii']
group_wii_mean=group_by_wii.Global_Sales.mean()
group_wii_mean
```
